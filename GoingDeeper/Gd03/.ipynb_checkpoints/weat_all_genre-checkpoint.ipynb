## STEP 1. 형태소 분석기를 이용하여 품사가 명사인 경우 해당 단어를 추출하기

```python
import os

data_dir = os.path.join(os.getenv("HOME"), "work/weat/data")
file_name = os.path.join(data_dir, "synopsis.txt")

with open(file_name, 'r') as file:
    for i in range(20):
        print(file.readline(), end='')
```

```python
# 약 15분정도 걸립니다.
from konlpy.tag import Okt

okt = Okt()
tokenized = []

with open(file_name, 'r') as file:
    while True:
        line = file.readline()
        if not line: break
        words = okt.pos(line, stem=True, norm=True)
        res = []
        for w in words:
            if w[1] in ["Noun"]:      # "Adjective", "Verb" 등을 포함할 수도 있습니다.
                res.append(w[0])    # 명사일 때만 tokenized 에 저장하게 됩니다.
        tokenized.append(res)

print("슝~")
```

## STEP 3. target, attribute 단어 셋 만들기

### TF-IDF로 해당 데이터를 가장 잘 표현하는 단어 셋 만들기

### (1) 영화별 대표단어 추출

```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from konlpy.tag import Okt

art_txt = 'synopsis_art.txt'
gen_txt = 'synopsis_gen.txt'

def read_token(file_name):
    okt = Okt()
    result = []
    with open(data_dir + '/' + file_name, 'r') as fread:
        print(file_name, '파일을 읽고 있습니다.')
        while True:
            line = fread.readline()
            if not line: break
            tokenlist = okt.pos(line, stem=True, norm=True)
            for word in tokenlist:
                if word[1] in ["Noun"]:#, "Adjective", "Verb"]:
                    result.append((word[0]))
    return ' '.join(result)

print("슝~")
```

```python
# 2개의 파일을 처리하는데 10분 가량 걸립니다.
art = read_token(art_txt)
gen = read_token(gen_txt)
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform([art, gen])

print(X.shape)
```

```python
print(vectorizer.vocabulary_['영화'])
print(vectorizer.get_feature_names_out()[23976])
```

```python
m1 = X[0].tocoo()   # art를 TF-IDF로 표현한 sparse matrix를 가져옵니다.
m2 = X[1].tocoo()   # gen을 TF-IDF로 표현한 sparse matrix를 가져옵니다.

w1 = [[i, j] for i, j in zip(m1.col, m1.data)]
w2 = [[i, j] for i, j in zip(m2.col, m2.data)]

w1.sort(key=lambda x: x[1], reverse=True)   #art를 구성하는 단어들을 TF-IDF가 높은 순으로 정렬합니다.
w2.sort(key=lambda x: x[1], reverse=True)   #gen을 구성하는 단어들을 TF-IDF가 높은 순으로 정렬합니다.

print('예술영화를 대표하는 단어들:')
for i in range(100):
    print(vectorizer.get_feature_names_out()[w1[i][0]], end=', ')

print('\n')

print('일반영화를 대표하는 단어들:')
for i in range(100):
    print(vectorizer.get_feature_names_out()[w2[i][0]], end=', ')
```

```python
n = 15
w1_, w2_ = [], []
for i in range(100):
    w1_.append(vectorizer.get_feature_names_out()[w1[i][0]])
    w2_.append(vectorizer.get_feature_names_out()[w2[i][0]])

# w1에만 있고 w2에는 없는, 예술영화를 잘 대표하는 단어를 15개 추출한다.
target_art, target_gen = [], []
for i in range(100):
    if (w1_[i] not in w2_) and (w1_[i] in model.wv): target_art.append(w1_[i])
    if len(target_art) == n: break

# w2에만 있고 w1에는 없는, 일반영화를 잘 대표하는 단어를 15개 추출한다.
for i in range(100):
    if (w2_[i] not in w1_) and (w2_[i] in model.wv): target_gen.append(w2_[i])
    if len(target_gen) == n: break
```

## STEP 4. WEAT score 계산과 시각화

### (1) embedding model과 단어 셋으로 WEAT score 구해보기

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

print("슝~")
```

```python
def cos_sim(i, j):
    return dot(i, j.T)/(norm(i)*norm(j))

def s(w, A, B):
    c_a = cos_sim(w, A)
    c_b = cos_sim(w, B)
    mean_A = np.mean(c_a, axis=-1)
    mean_B = np.mean(c_b, axis=-1)
    return mean_A - mean_B #, c_a, c_b

def weat_score(X, Y, A, B):

    s_X = s(X, A, B)
    s_Y = s(Y, A, B)

    mean_X = np.mean(s_X)
    mean_Y = np.mean(s_Y)

    std_dev = np.std(np.concatenate([s_X, s_Y], axis=0))

    return  (mean_X-mean_Y)/std_dev
```

```python
matrix = [[0 for _ in range(len(genre_name))] for _ in range(len(genre_name))]
print("슝~")
```

```python
X = np.array([model.wv[word] for word in target_art])
Y = np.array([model.wv[word] for word in target_gen])

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        A = np.array([model.wv[word] for word in attributes[i]])
        B = np.array([model.wv[word] for word in attributes[j]])
        matrix[i][j] = weat_score(X, Y, A, B)

print("슝~")
```

## STEP 5. 추가 실험

### 장르별 단어 중복을 없이 결과 도출
- 전체 장르에서 중복을 없애는 것은 아님
- 장르A 와 장르B의 weat score를 구할 때 A,B 간의 중복 없는 상위 15개를 attribute로 사용

```python
## 장르별 명사 다시 준비

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(genre)

print(X.shape)
```

```python
## 중복 없는 15개 대표 단어를 뽑기 위해 상위 빈도 100개씩을 준비

m = [X[i].tocoo() for i in range(X.shape[0])]

w = [[[i, j] for i, j in zip(mm.col, mm.data)] for mm in m]

for i in range(len(w)):
    w[i].sort(key=lambda x: x[1], reverse=True)
attributes = []
for i in range(len(w)):
    print(genre_name[i], end=': ')
    attr = []
    j = 0
    while (len(attr) < 100):
        if vectorizer.get_feature_names_out()[w[i][j][0]] in model.wv:
            attr.append(vectorizer.get_feature_names_out()[w[i][j][0]])
            print(vectorizer.get_feature_names_out()[w[i][j][0]], end=', ')
        j += 1
    attributes.append(attr)
    print('\n')
```

```python
# 각 장르별로 고유한 단어 15개씩 추출하여 WEAT 스코어 계산

def get_unique_genre_words(genre_i, genre_j, attributes_all, model, n=15):
    """
    두 장르에서 서로 중복되지 않는 고유 단어들을 추출
    
    Args:
        genre_i, genre_j: 장르 인덱스
        attributes_all: 전체 장르별 단어 리스트 (빈도순 정렬된 상태)
        model: Word2Vec 모델
        n: 추출할 단어 개수
    
    Returns:
        genre_i_unique, genre_j_unique: 각각의 고유 단어 리스트
    """
    
    # 각 장르의 전체 단어 리스트 (빈도순)
    words_i = attributes_all[genre_i]  # 장르 i의 모든 단어들 (빈도순)
    words_j = attributes_all[genre_j]  # 장르 j의 모든 단어들 (빈도순)
    
    # 장르 i에만 있고 장르 j에는 없는 단어들
    genre_i_unique = []
    for word in words_i:
        if (word not in words_j) and (word in model.wv):
            genre_i_unique.append(word)
            if len(genre_i_unique) == n:
                break
    
    # 장르 j에만 있고 장르 i에는 없는 단어들  
    genre_j_unique = []
    for word in words_j:
        if (word not in words_i) and (word in model.wv):
            genre_j_unique.append(word)
            if len(genre_j_unique) == n:
                break
                
    return genre_i_unique, genre_j_unique
```

```python
# 수정된 WEAT 스코어 계산 코드
X = np.array([model.wv[word] for word in target_art])  # 예술영화 단어들
Y = np.array([model.wv[word] for word in target_gen])  # 일반영화 단어들
```

```python
matrix_unique = [[0 for _ in range(len(genre_name))] for _ in range(len(genre_name))]
```

```python
un_duplicate_weat = []
categories = []

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        
        # 각 장르 쌍에서 서로 중복되지 않는 고유 단어들 추출
        genre_i_words, genre_j_words = get_unique_genre_words(i, j, attributes, model, n=15)
        
        # 충분한 단어가 없는 경우 스킵
        if len(genre_i_words) < 15 or len(genre_j_words) < 15:
            print(f"경고: {genre_name[i]} vs {genre_name[j]} - 충분한 고유 단어가 없습니다.")
            print(f"  {genre_name[i]}: {len(genre_i_words)}개, {genre_name[j]}: {len(genre_j_words)}개")
            continue
            
        # 임베딩 벡터로 변환
        A = np.array([model.wv[word] for word in genre_i_words])
        B = np.array([model.wv[word] for word in genre_j_words])
        
        # WEAT 스코어 계산
        matrix_unique[i][j] = weat_score(X, Y, A, B)
        
        # 결과 출력 (선택사항)
        print(f"{genre_name[i]} vs {genre_name[j]}: {matrix_unique[i][j]:.3f}")
        print(f"  {genre_name[i]} 고유단어: {genre_i_words[:15]}...")  # 처음 5개만 출력
        print(f"  {genre_name[j]} 고유단어: {genre_j_words[:15]}...")
        # print()

        categories.append(f"{genre_name[i]} vs {genre_name[j]}")
        un_duplicate_weat.append(matrix_unique[i][j])
        

print("완료!")
```

```python
# 시각화 (기존과 동일)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(17, 17))
ax = sns.heatmap(matrix_unique, 
                 xticklabels=genre_name, 
                 yticklabels=genre_name, 
                 annot=True, 
                 cmap='RdYlGn_r',
                 # fmt='.3f'
                )
plt.title('장르별 고유 단어')
```

### 느낀점
- 편향된 세상의 데이터를 ai 엔지니어가 바꾸려고 하면
- 사용자 입장에서는 오히려 결과(서비스)가 이상하다고 생각될 수 있음
- 왜냐면 이미 사용자의 생각이 편향되어 있기 때문
- 그래서 편향의 문제는 ai 엔지니어가 바로 잡는 것이 아니라 세상이 바로 잡아서 편향 없는 데이터를 만들어야 한다고 생각함
- 그렇지 않으면 흑어공주(중세 덴마크에 흑인이...), 구릿빛 백설공주(피부가 하얘서 백설인데) 이런 문제가 만들어짐
