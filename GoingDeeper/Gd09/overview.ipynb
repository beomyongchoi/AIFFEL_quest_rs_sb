# KoChatGPT 업그레이드 하기

## 프로젝트 개요
- Pretrained 모델에 RLHF 적용하여 성능을 증가 시킨다는 기본 전제
- 어떻게하면 성능을 올릴 수 있을가 고민해 봄
- KoChatGPT 구현한 저자가 data를 생성한 방법대로 새로운 데이터를 생성하기로 함
- 저자는 '몇년전' ChatGPT를 이용하여 data를 생성하였으므로 더 발전된 ChatGPT 모델로 새 데이터를 생성하기로 함
- gtp-4o를 사용하려 했으니 응답이 오래 걸리고 프로젝트 기간의 압박으로 gpt-4o-mini를 사용함
- openAI api를 사용하여 SFT, RM data를 생성

### SFT data
- prompt를 입력으로 하여 나오는 출력으로 기존 completion을 대체
```
 SYSTEM_PROMPT = """당신은 도움이 되고 친근한 한국어 AI 어시스턴트입니다. 
사용자의 질문에 정확하고 유용한 답변을 제공해주세요. 
답변은 자연스럽고 이해하기 쉽게 작성해주세요."""
```

### RM data
- prompt를 입력으로 하여 나오는 출력으로 기존 completion 3개에 4개를 더 추가
- 새로 생성된 completion을 ranking 0가 되도록 처리

```
SYSTEM_PROMPT = """당신은 정확하고 도움이 되는 한국어 AI 어시스턴트입니다. 
사용자의 질문에 간결하고 정확한 답변을 제공해주세요. 
불확실한 정보에 대해서는 추측하지 말고 모른다고 답하세요."""
```

- base: 노드에 있는 내용으로 base 코드 만듬

- changed_model_new_data: changed_model + new_dat
- changed_model_new_data_increased_epoch: changed_model_new_data에 epoch 수를 3으로 증가 시킴
